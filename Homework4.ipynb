```python
import cv2 
import numpy as np
from matplotlib import pyplot as plt

plt.rcParams['figure.figsize'] = (20, 10)
```

```python
def pltImage(img):
    channels = len(img.shape)
    if channels == 3:
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    else:
        plt.imshow(img, cmap='gray')
    plt.xticks([]), plt.yticks([]) # to hide tick values on X and Y axis
    plt.show()
```

```python
img = cv2.imread('dsu7.jpg') 
rows,cols,ch = img.shape
pts1 = np.float32([[250,270],[570,250],[570,500],[240,462]])
pts2 = np.float32([[250,250],[500,250],[500,400],[250,400]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(1025,770))


plt.subplot(121),plt.imshow(dst),plt.title('Input')
plt.subplot(122),plt.imshow(img),plt.title('Output')
plt.show()
```
